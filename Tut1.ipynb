```r
# Import data
MyData <- read.csv(file="../input/A1.csv", header=TRUE, sep=",")
# View data
data
```

```r
plot(MyData$height, MyData$head, main='Scatterplot of Head Circumference and Heigh (8350)',
     xlab='Heigh (cm)', ylab='Head Circumference (cm)', col = 'red' )
```

```r
summary(MyData$height)
MyData$height
mean(MyData$height)
sd(MyData$height)
var(MyData$height)
IQR(MyData$height)
par(yaxs="i",las=1)
hist(MyData$height,
prob=TRUE,col="black",border="white",
xlab="Heigh (cm)",
main="Histrogram of Height (8350)")
box(bty="l")
lines(density(MyData$height,na.rm=T),col="red",lwd=4)
grid(nx=NA,ny=NULL,lty=1,lwd=1,col="gray")
```

```r
summary(MyData$head)
mean(MyData$head)
sd(MyData$head)
var(MyData$head)

par(yaxs="i",las=1)
hist(MyData$head,
prob=TRUE,col="black",border="white",
xlab="Head Circumference (cm)",
main="Histrogram of Head Circumference (8350)")
box(bty="l")
lines(density(MyData$head,na.rm=T),col="red",lwd=4)
grid(nx=NA,ny=NULL,lty=1,lwd=1,col="gray")
MyData$head
```

```r
mod=lm(head ~ height, MyData)
summary(mod)
plot(head ~ height, data = MyData,  main= 'Linear Regression Fit for Head Circumference and Height (8350)', xlab='Height (cm)', ylab='Head Circumference (cm)', col = 'red')
abline(mod, col='blue')
```

```r
plot(MyData$height, resid(mod), main='Residual Plot (8350)', xlab = 'Height (cm)', ylab='Residuals')

summary(resid(mod))
IQR(resid(mod))
abline(0,0)
```
