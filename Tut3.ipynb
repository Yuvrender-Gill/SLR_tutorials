```r
data <- read.csv(file="../input/reale_a3data.csv", sep=",")
head(data)
```

## Removing the NA values from the data and then getting new dimensions

```r
data <- na.omit(data)
dim(data)
```

```r
data$lotsize<- data$lotwidth*data$lotlength
dim(data)
```

```r
corr_matrix <- cor(data[c(2, 3, 4, 5, 6, 7, 8, 9, 11)] )
round(corr_matrix, 4)
```

```r
library(GGally)
ggpairs(data[c(2, 3, 4, 5, 6, 7, 8, 9, 11)], upper = list(continuous = "points", combo =
  "facethist", discrete = "facetbar", na = "na") )
```

Q2

```r
plot(data$lotlength, data$Sale)
```

```r
slr_model <-lm(Sale~lotlength, data=data)
summary(slr_model)
```

```r
plot(slr_model)
```

```r

df <- data.frame(y=1+2*data$lotlength+rnorm(162)/2,  wght1=1:162)

## Fancy weights as numeric vector
wht_slr_model <- lm(data$Sale~data$lotlength,data=df,weights=(df$wght1)^(3/4))
summary(wht_slr_model) 
```

```r
plot(wht_slr_model)
```

Q3

```r
mlr_model <- lm(Sale ~ list + bedrooms + bathrooms + lotwidth + lotlength + maxsqfoot + taxes + location + lotsize, data=data)
```

```r
summary(mlr_model)
```

```r
round(summary(mlr_model)$coefficient[, c(1,3,4)], 4)
```
